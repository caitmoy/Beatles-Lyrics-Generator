## Step 2: Web Scraper

After selecting the corpora, the next step was to format it and prep the text for use.

For each URL, I used BeautifulSoup (HTML parser) to help grab the text that I needed. Using regular expressions, I marked newline tokens with "#" to maintain where the Beatles would have naturally broken up lines. I also wrote a helper method to trim the website content, making sure to only keep the lyrics and discarding any unrecessary text. Regular expressions were also used to make stylistic changes to the text. I changed "wo" to "woah", "yeh" to "yeah", and capitalized "i" purely for preference, and to help the lyrics read more naturally. After each song was formatted, it was written as a string to the "lyrics.txt" file created earlier. 

```python
import requests
from bs4 import BeautifulSoup
import re

songs = []

# helper method to trim website content to just include lyrics
def trimBody(text):
    lyrics = []
    index = 0
    for word in text:
        # after you seeing Lennon & Mccartney, add the lyrics to a list
        if (word == 'McCartney'):
            index = index + 1
            text = text[index:]
            index = 0
        # after seeing "Visit also..." get rid of text following the lyrics
        if (word == 'Visit'):
            text = text[:index-1]
        index = index + 1
    
    # now we are left with only lyrics
    # return a string with spaces between words
    return " ".join(text)

# go to each url and get only the lyrics and format it for use
for url in sites:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    body = soup.body.text
    body = re.findall(r'\S+|\n',body) # keep new lines
    trimmed_content = trimBody(body).lower()
    
    # add '#' symbol at start of each line and end of each line
    trimmed_content = re.sub(r'\n', '#',trimmed_content)
    trimmed_content = re.sub(r'((# )+)|(( #)+)'," # ",trimmed_content)
          
    #stylistic changes
    trimmed_content = re.sub(r'\bi\b',"I",trimmed_content)
    trimmed_content = re.sub(r'\b(wo)\b',"woah",trimmed_content)
    trimmed_content = re.sub(r'\b(yeh)\b',"yeah", trimmed_content)
    trimmed_content = re.sub(r'[\(\)]',"", trimmed_content)

    songs.append(trimmed_content) # store lyrics in songs array

# write all songs to text file
for song in songs:
    with open("lyrics.txt", "a") as text_file:
        text_file.write(song + '\n')
```

## Step 3: Bigrams

Due to the small sample size of the corpora, bigrams yielded the best results. I used the bigrams() and bigram_at() functions from class, changing them only to keep punctuation to better perserve the feel of the Beatles' lyrical style. The frequency of each bigram was then counted, and the bigrams were ordered from most frequently occurring to least frequently occurring (mostly for testing purposes). I also did some error-checking for unwanted bigrams. 

```python
import operator
import re

def bigram_at (l,i):
    t = (l[i], l[i+1])
    return t

def bigrams(s):
    tokens = re.findall(r"[^-\s]+",s) #keep punctuation in
    allbigrams = []
    i = 0
    while i < len(tokens)-1:
        allbigrams.append(bigram_at(tokens,i))
        i += 1
    return allbigrams

# open text file as string
with open('lyrics.txt', 'r') as file:
    data = file.read()
    
# make into bigrams
all_bigrams = bigrams(data)

# count the frequency of each bigram
counts = dict()
for bigram in all_bigrams:
    if bigram in counts:
        counts[bigram] += 1
    else:
        counts[bigram] = 1

# sort the bigram frequencies from most occurrences to least occurrences       
sorted_counts = dict(sorted(counts.items(), key=operator.itemgetter(1),reverse=True))

# double check to get rid of unwanted bigrams
sorted_counts.pop(('#', '#'), None)
sorted_counts.pop(('I', 'I'), None)    
```

## Step 4: Text Generation

I wrote the method generate_verse() to deal with text generation. This method creates a 50-word phrase of Beatles lyrics given a starting word. It uses the helper method starts_with() to find all bigrams that begin with the starting word. From those options, it selects a random bigram, taking into account the weighted probabilities. If no bigrams exist with the starting word, a random bigram is picked, again, taking into account the weighted probabilities. Once a bigram is selected, the first word of the bigram is printed.

This method also keeps track of how many words there are per line. If a line gets too long (>15 words), a new line is created. If the bigram is an ending phrase (word, #), a new line is created. If the bigram is a starting phrase (#, word), the function makes sure not to print the # symbol. 


```python
import random

# helper method
def starts_with(t,bigs):
    r = {}
    for i in bigs:
        if i[0] == t:
            r[i] = bigs.get(i)
    return r

def generate_verse(starting_word):
    words_per_line = 0

    for x in range(50):
        options = []
        # get options from provided starting word
        options = starts_with(starting_word, sorted_counts)
        
        # if no options, pick random word
        if len(options) < 1:
            selected_bigram = random.choices(list(sorted_counts.keys()),weights=sorted_counts.values(), k=1)
        # pick a random word using weighted probabilties of possible options
        else:
            selected_bigram = random.choices(list(options.keys()), weights=options.values(), k=1)
        
        # if too many words per line, break it up
        if (words_per_line > 15):
            print()
            words_per_line = 0
        
        # if bigram is an ending phrase (word, #), print the word and start a new line
        if (selected_bigram[0][1] == "#"):
            print(selected_bigram[0][0])
            words_per_line = 0 # reset 
            
        # all other cases; skip instances of (#, word)- this saves the word, but doesn't print the #
        elif (selected_bigram[0][0] != "#"):    
            print(selected_bigram[0][0],end=" ")

        # new starting word
        starting_word = selected_bigram[0][1]
        words_per_line += 1
```
